From Collection to Presentation

```python
# import necessary packages

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# loading data

root_dir = "../data/IMF/climate-change-data/"
file_name = "4_CO2_Emissions_Emissions_Intensities_and_Emissions_Multipliers.csv"
```

```python
df = pd.read_csv(root_dir + file_name)
```

```python
# let's check the few rows of the dataset
df.head()
```

```python
# let's check the statistical summary, number of row and column for numerical variable with describe
df.describe()
```

```python
# let's check missing values in each column
df.isnull().sum()
```

```python
years = list(map(str, range(1995, 2019)))
print(years)
```

```python
years_1 = list(['1995', '1996', '1997', '1998', '1999', '2000'])
years_2 = list(['2013', '2014', '2015', '2016', '2017', '2018'])
print(['Industry'] + years_1 + years_2)
```

```python
# group by industry
df_industry = df[["Industry"] + years_1 + years_2]
```

```python
df_industry.head()
```

```python
df_industry_gp = df_industry.groupby("Industry", as_index=False).mean()
df_industry_gp
```

```python
df_industry_gp = df_industry_gp.sort_values(years_1 + years_2, ascending=False)
```

```python
df_industry_gp.head()
```

```python
top_industry = df_industry_gp.head(10)
```

```python
top_industry.set_index('Industry', inplace=True)
```

```python
top_industry.index.name = None
```

```python
top_industry
```

```python
top_industry = top_industry.transpose()
top_industry
```

```python
# Set the index column as x-axis data
x = top_industry.index

# Get the column names for the y-axis
y_columns = top_industry.columns[:]  # Exclude the index column

# Create a line chart with multiple lines
plt.figure(figsize=(12, 6))
for col in y_columns:
    y = top_industry[col]
    plt.plot(x, y, label=col, marker='o')

# Customize the chart
plt.title('Line Chart of Values Over Time')
plt.xlabel('Year')
plt.ylabel('Millions of Metric tons of CO2')
plt.legend()

# Show the chart
plt.grid(True)
plt.show()
```

```python
# group by country
df_country = df[["Country"] + years_1 + years_2]
df_country
```

```python
# group by country
df_country_gp = df_country.groupby(["Country"], as_index=False).mean()
df_country_gp.head()
```

```python
# order by quantity emmited
df_country_sv = df_country_gp.sort_values(years_1 + years_2, ascending=False)
df_country_sv
```

```python
top_country = df_country_sv.head(10)
top_country
```

```python
top_country.set_index("Country", inplace=True)
top_country.index.name = None
top_country
```

```python
top_country = top_country.transpose()
top_country
```

```python
# Set the index column as x-axis data
x = top_country.index

# Get the column names for the y-axis
y_columns = top_country.columns[:]

# Create a line chart with multiple lines
plt.figure(figsize=(12, 6))
for col in y_columns:
    y = top_country[col]
    plt.plot(x, y, label=col, marker='o')

# Customize the chart
plt.title('CO2 Emmission per Country')
plt.xlabel('Year')
plt.ylabel('Millions of Metric tons of CO2')
plt.legend()

# Show the chart
plt.grid(True)
plt.show()
```
